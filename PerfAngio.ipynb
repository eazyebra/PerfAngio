```python
# import libraries 
from scipy.io import loadmat 
import matplotlib.pyplot as plt
import numpy as np
```

```python
# upload mat file
data = loadmat('perfAngio_130.mat')
#print(data.keys())
img_3d = data['X'] # access data from X key

# display each 2D slice 
for i in range(20):
    slice_2d = img_3d[:, :, i]
    plt.figure()
    plt.imshow(slice_2d, cmap='gray')

    
    
```

```python
# randomly sample p number of pixels 
slice_2d = img_3d[:, :, 1]
print(np.shape(slice_2d)) # 1024 x 1024

def random_pixels(p):
    size = (p,2)
    rand_pixels = np.zeros(size)
    for i in range(p):
        random_index1 = np.random.randint(0, slice_2d.shape[0]) # select randint from 0 to x dim (1024)
        random_index2 = np.random.randint(0, slice_2d.shape[1]) # select randint from 0 to y dim (1024)
        rand_pixels[i] = slice_2d[random_index1, random_index2] # index pixel values, store in rand_pixels
    return rand_pixels
        

random_pixels(5)

```
